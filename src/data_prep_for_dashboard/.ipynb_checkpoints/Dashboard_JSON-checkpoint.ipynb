```python
import json
import csv
import pandas as pd
```

```python
df = pd.read_csv('all_agts_with_China.csv')
print("Columns in the CSV file:", df.columns.tolist())
```

```python
def process_csv(file_name, output_file_base):
    
    df = pd.read_csv(file_name)
    
    network_df = df[['AgtId', 'agt_dat', 'short_agt_name', 'agt_description', 'PPName', 'date', 'stage_label', 'agt_type', 'year', 'signatory_type', 'actor_name', 'actor_type', 'PAX_Hyperlink']]
    
    network_df['agt_dat'] = network_df['agt_dat'].str.replace(r' \(\d{2}\/\d{2}\/\d{4}\)', '', regex=True)
    
    network_df['date'] = pd.to_datetime(network_df['date'])
    network_df['date'] = network_df['date'].dt.strftime('%d/%m/%Y')

    network_df.rename(columns={'agt_description': 'description'}, inplace=True)
    network_df.rename(columns={'agt_dat': 'Agt'}, inplace=True)
    
    
    network_df['Type'] = 'Peace Agreement'
    network_df['target_type'] = 'signatory'
    
    csv_output_file = output_file_base + '.csv'
    network_df.to_csv(csv_output_file, index=False)
    
    json_output_file = output_file_base + '.json'
    network_df.to_json(json_output_file, orient='records', indent=4)

files_to_process = ['all_agts_with_United_Kingdom.csv', 'all_agts_with_China.csv', 'all_agts_with_Russia.csv']

# Process each file
for file in files_to_process:
    base_name = 'modified_' + file.split('.')[0]
    process_csv(file, base_name)
```
