```python
# Initiate
import os
import pandas as pd
```

```python
# Source data file path
link_table_path = 'all_agts_with_UK.csv'

# Read the CSV
try:
    df = pd.read_csv(link_table_path)
    
    # Now you can work with the DataFrame 'df'
    print("First few rows of the DataFrame:")
    print(df.head(2))
    
    # access any column
    print("Values in the 'column_name' column:")
    print(df['signatory_type'])
    
except FileNotFoundError:
    print("CSV file not found at the specified path.")
```

```python
# List all fields for creating networks
columns_to_network = [
    'AgtId', 
    'Agt', 
    'short_agt_name', 
    "description", 
    "Type", 
    "PPName", 
    "agt_type", 
    "stage_label", 
    "date", 
    "year", 
    "signatory_type", 
    "actor_name", 
    "new_actor_type", 
    "additional_info", 
    "PAX_Hyperlink"
]

# Create a new DataFrame for network
network_df = df[columns_to_network].copy()

# Add a new column "signatory_type" with the value "signatory"
network_df['target_type'] = 'signatory'

print("New DataFrame with selected columns:")
print(network_df.head(2))
```

```python
# Export to CSV
network_df.to_csv('network_data.csv', index=False)

# Export to JSON
# network_df.to_json('network_data.json', orient='records', lines=True)
```
